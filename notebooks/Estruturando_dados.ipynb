```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
path = '/dados/GoogleDrive/rotinas_python/github_projects/hmdbwebscraping/notebooks/data/'
filenames = sorted([path+name for name in os.listdir(path)])
filenames
```

```python
colnames = ['Name','Smiles','Inchi','CasRN','MeltingPoint',
                         'BoilingPoint','WaterSolubility_mg_mL','LogP','Ref',
                         'ppm_exp','n_peaks_exp','multiplicity_exp','n_Hs_exp','atom_number_exp','peak_centers_exp','ExperimentalConditions',
                         'ppm_pred','n_peaks_pred','multiplicity_pred','n_Hs_pred','atom_number_pred','peak_centers_pred','predCondition']
df_full = pd.DataFrame(columns=colnames)

for path in tqdm(filenames):
    df_iter = pd.read_csv(path,sep=';')
    df_full = pd.concat([df_full,df_iter], axis=0)

df_full = df_full.reset_index(drop=True)
```
